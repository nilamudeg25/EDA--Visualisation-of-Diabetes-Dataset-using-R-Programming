# The objective of the dataset is to predict whether a patient has diabetes or not, based on certain diagnostic measurements included in the dataset.

The different columns present in the dataset are:

Pregnancies -> Number of times Pregnant

Glucose -> Plasma glucose concentration

BloodPressure -> Diastolic blood pressure (mm Hg)

SkinThickness -> Triceps skin fold thickness (mm)

Insulin -> 2-Hour serum insulin (mu U/ml)

BMI -> Body Mass Index

DiabetesPedigreeFunction -> Diabetes pedigree function

Age -> Age in years

Outcome -> Whether the lady is diabetic or not, 0 represents the person is not diabetic and 1 represents that the person is diabetic.

#Box Plot : create a boxplot by using geom_boxplot() and it is used to display the outliers 


```r
#outliers in the BMI feature 

p1=ggplot(data=df,aes(x=Outcome,y=BMI,fill=Outcome))

box_plot1<-p1+geom_boxplot()
print(box_plot1)


```

1) BMI levels are >20 and < 35 then, there is very low probability of having diabetes

2) BMI levels are > 30 and < 40 then, there is high probability of having diabetes

```r
# outliers in the Glucose feature data

p2=ggplot(data=df,aes(x=Outcome,y=Glucose,fill=Outcome))

box_plot2<-p2+geom_boxplot()
print(box_plot2)
```

1) Glucose levels are between 90-125 then there is very low probability of having diabetes

2) Glucose levels are between 125-175 then there is high  probability of having diabetes

#Histogram : A histogram represents the frequencies of values of a variable bucketed into ranges. R creates histogram using hist() function.

#Scatter Plot : To see the relationship between two numeric variables used scatter plot.

geom point : The function geom_point() adds a layer of points in a plot, which creates a scatterplot.



```r
# The relationship between Age and Glucose

p3=ggplot(data=df,aes(x=Age,y=Glucose,col=Outcome))+geom_point()
```

#geom smooth : geom_smooth() adds a trend line over an existing plot.

```r
p3+geom_smooth()  # plot average line of each category 
```

1) 0: Glucose levels are between 100-140 then there is very low probability of having diabetes

2) 1: Glucose levels are between 145-175 then there is high probability of having diabetes


```r
# The relationship between Age and BMI
p4=ggplot(data=df,aes(x=Age,y=BMI))+geom_point()+geom_smooth(aes(col=Outcome))
print(p4)
```

1) 0: The BMI level is decreases of those who don't have a diabetes.

2) 1: The BMI level is increases of those who have a diabetes.


# A 2d density plot is useful to see the relationship between 2 numeric variables

# A Density Plot visualises the distribution of data over a continuous interval.

```r
# The graph between AGE vs BMI

ggplot(data = df, mapping = aes(x = Age, y = BMI)) +
  geom_bin2d() +
  geom_density_2d() +
  geom_point()
```

It tells for perticular density how many counts are there. Black means lower density and blue means higher density.

```r
# AGE vs BLOODPRESSURE

ggplot(data = df, mapping = aes(x = Age, y = BloodPressure)) +
  geom_density_2d() +
  geom_point()
```

```r
# relationship between AGE and GLUCOSE

ggplot(data = df, mapping = aes(x = Age, y = Glucose)) + geom_point(alpha = 0.1, color = "blue", size = 5)+ geom_density_2d(color = "red") 
```

# faceting : Generated by number of variables. Distinguish row wise & column wise. In facet we got each graph separately so we can analyse each graph.

```r
ggplot(data = df, mapping = aes(x = Age, y = Glucose)) +
  geom_point() +
  geom_smooth(aes(color = Outcome)) +
  facet_wrap( ~ Outcome)
```

1) non-diabetes(0)= the people who dont have diabetes but they have glucose.

2) diabetes(1) = the people who have diabetes they also have glucose.


```r

ggplot(data = df, mapping = aes(x = Age, y = BMI)) +
  geom_point() +  # used to create scatter plot
  geom_smooth(aes(color = Outcome)) +   # trend the line on existing plot
  facet_wrap(~Outcome)  # generated by number of variables here 2 variables are there 0,1 
```

1) non-diabetes(0)= the people who dont have diabetes they have less BMI between 20-30

2) diabetes(1) = the people who have diabetes and they are overweight
